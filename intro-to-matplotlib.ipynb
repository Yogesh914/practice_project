# Introduction to Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
plt.plot();
```

```python
x = [1,2,3,4]
y = [4,56,42,6]

plt.plot(x,y);
```

```python
# 1st method of plotting 

fig = plt.figure() # creates figure
ax = fig.add_subplot() # adds axes

plt.show();
```

```python
# 2nd method

fig = plt.figure()
ax = plt.axes()

ax.plot(x,y)
plt.show();
```

```python
#3rd method *recommended

fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig, ax = plt.subplots(figsize = (10,10)) # (width, height)

ax.plot(x,y)

#Customizing your plot
ax.set(title="Sample Plot", xlabel="x-axis", ylabel="y-axis")

#saving plot
fig.savefig("images/sample_plot.png")
```

```python
# Subplots method 1

fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

# plotting on the subplots

ax1.scatter(x, y)
ax2.plot(x, np.sin(x))
ax3.plot(x, x**3);
```

```python
# 2nd method subplots

fig, ax = plt.subplots(nrows=2, ncols =2)

ax[0,0].plot(x, x**3)
ax[1,0].plot(x, y)
ax[1,1].hist(np.random.randn(1000));
```

## Two methods

* pyplot method for quick and easy visualizations 
* matplotlib OO method for more advanced

```python
over_60 = heart_data[heart_data["age"] > 60]
over_60
```

```python
#pyplot method

over_60.plot.scatter(x="age", y="chol", c="target");
```

```python
#OO method

fig, ax = plt.subplots()

over_60.plot.scatter(x="age", y="chol", c="target", ax=ax);

# ax.set_xlim([40, 80])
```

```python
#3 OO method from scratch

fig, ax = plt.subplots()

scatter = ax.scatter(x=over_60["age"], y=over_60["chol"], c=over_60["target"])

ax.set(title="Heart Disease with age and cholesterol levels", xlabel="Age", ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_60["chol"].mean(), ls='--'); #ls = linestyle
```

```python
over_60.head()
```

```python
# subplots

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(15,10))



scatter = ax1.scatter(x=over_60["age"], y=over_60["chol"], c=over_60["target"])

ax1.set(title="Heart Disease with age and cholesterol levels", ylabel="Cholesterol")

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y=over_60["chol"].mean(), ls='--');



scatter = ax2.scatter(x=over_60["age"], y=over_60["thalach"], c=over_60["target"])

ax2.set(title="Heart Disease with age and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

ax2.legend(*scatter.legend_elements(), title="Target")

ax2.axhline(y=over_60["thalach"].mean(), ls='--');


fig.suptitle("Heart Disease Analysis", fontweight="bold", fontsize=20);
```
